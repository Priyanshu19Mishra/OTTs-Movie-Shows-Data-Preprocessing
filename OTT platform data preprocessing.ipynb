Genre visualization and analysis

```python
data_Genre=data.copy()
del data_Genre['listed_in']
s = data['listed_in'].str.split(',').apply(pd.Series, 1).stack()
s.index = s.index.droplevel(-1)
s.name = 'listed_in'
data_Genre = data_Genre.join(s)
```

```python
data_Genre.head()
```

Clustered Analysis due to different genres that are present in a single Coloumn

```python
genre_counts = data['listed_in'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.xticks(rotation=45)
plt.xlabel('listed_in')
plt.ylabel('Number of Content')
plt.title('Distribution of Content by Genre')
plt.tight_layout()
plt.show()
```

Implementing the frequnecy of each genre from all genres using "Word_cloud". This shows the most liked genre across all the movies in the dataset. This basically shows that the OTT platforms are more interested to buy these genres which typliaclly shows the choice of Genres among the audiences.

```python
lis=[]
for i in data_Genre['listed_in']:
    lis.append(i)

for k in range(0,len(lis)):
    lis[k]=str(lis[k]).strip()

from collections import Counter
genre_count = Counter(lis)

from wordcloud import WordCloud
wc = WordCloud(background_color='white')
wc.generate_from_frequencies(genre_count)
plt.figure(figsize=(12,8))
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

Indiviual Genre analysis under different platforms

```python
fig,axes=plt.subplots(1,2,figsize=(15,5))
data_Genre.loc[data_Genre['Netflix_data']==1,'listed_in'].value_counts().plot(kind='bar',ax=axes[0],color='orangered')
data_Genre.loc[data_Genre['Hulu_data']==1,'listed_in'].value_counts().plot(kind='bar',ax=axes[1],color='forestgreen')

axes[0].set_title('Netflix Genres')
axes[1].set_title('Hulu Genres')

plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(15,5))
data_Genre.loc[data_Genre['Amazon_data']==1,'listed_in'].value_counts().plot(kind='bar',ax=axes[0],color='indigo')
data_Genre.loc[data_Genre['Disney+Hotstar_data']==1,'listed_in'].value_counts().plot(kind='bar',ax=axes[1],color='crimson')

axes[0].set_title('Amazon Genres')
axes[1].set_title('Disney+ Genres')

plt.show()
```
